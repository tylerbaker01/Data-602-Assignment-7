### Data Wrangle
There are a few problems with this data table.
1. Unamed column is redundant.
2. The category column is all of the same value. Can just be dropped.
3. Melt levels into rows and turn #, % into columns.
4. Level"3+4+ isn't needed.

```python
# Drop columns
cols = ["Unnamed Column", "Category", "# Level 3+4", "% Level 3+4"]
math_df = math_df.drop(cols, 1)
math_df
```

```python
new_math = pd.melt(math_df, id_vars = ["DBN", "School Name", "Grade", "Year", "Number Tested", "Mean Scale Score", "% Level 1",  "% Level 2", "% Level 3", "% Level 4"], var_name = "Level", value_vars = ["# Level 1",  "# Level 2", "# Level 3", "# Level 4"], value_name="# of students")
new_math
```

```python
# rename columns to have better observations when I pivot.
new_math.rename(columns={"% Level 1": "1", "% Level 2": "2", "% Level 3": "3", "% Level 4": "4"}, inplace = True)
df_math = pd.melt(new_math, id_vars = ["DBN", "School Name", "Grade", "Year", "Number Tested", "Mean Scale Score", "# of students"], var_name = "Level", value_vars = ["1",  "2", "3", "4"], value_name = "% of students")
df_math
```
